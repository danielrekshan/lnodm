```python
dd = model.get_document_info(docs)
all = pd.concat([df, dd], axis=1, join="outer")
all

# df
```

```python
all['city'] = all['city'].apply(lambda x: str(x).title())
all['state'] = all['state'].apply(lambda x: str(x).title())
def country_codes(country):
    if country.title() == "Australia":
        return "AU"
    elif country.title() == "United States":
        return "US"
    elif country.title() == "United Kingdom":
        return "UK"
    elif country.title() == "Philippines":
        return "PH"
    elif country.title() == "Nicaragua":
        return "NI"
    elif country.title() == "Hong Kong":
        return "HK"
    elif country.title() == "Canada":
        return "CA"
    else :
        return country
all['country'] = all['country'].apply(lambda x: country_codes(str(x)))


```
